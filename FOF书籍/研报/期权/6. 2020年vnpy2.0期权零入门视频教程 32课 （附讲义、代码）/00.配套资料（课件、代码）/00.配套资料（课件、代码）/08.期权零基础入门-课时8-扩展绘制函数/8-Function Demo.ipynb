```python
import matplotlib.pyplot as plt
```

```python
def plot_call_price():
    k = 3000
    x = []
    y = []
    
    for s in range(2500, 4000, 100):
        p = max(s - k, 0)
        x.append(s)
        y.append(p)
        
    plt.plot(x, y)

plot_call_price()
```

```python
def plot_call_price(k=3000):
    x = []
    y = []
    
    for s in range(2500, 4000, 100):
        p = max(s - k, 0)
        x.append(s)
        y.append(p)
        
    plt.plot(x, y)

plot_call_price(3500)
```

```python
def plot_call_price(k=3000, start=2500, end=4000, step=100):
    x = []
    y = []
    
    for s in range(start, end, step):
        p = max(s - k, 0)
        x.append(s)
        y.append(p)
        
    plt.plot(x, y)

plot_call_price(3500, 3000, 5000, 10)
```

```python
def plot_option_price(k=3000, start=2500, end=4000, step=100, cp=1):
    x = []
    y = []
    
    for s in range(start, end, step):
        # if cp == 1:
        #     p = max(s - k, 0)
        # else:
        #     p = max(k - s, 0)
        p = max(cp * (s - k), 0)
        
        x.append(s)
        y.append(p)
        
    plt.plot(x, y)

plot_option_price(3500, 3000, 5000, 10, -1)
```

```python
plot_option_price(3500, 3000, 5000, 10, 1)
```

```python
def plot_option_price(k=3000, start=2500, end=4000, step=100, cp=1, direction=1):
    x = []
    y = []
    
    for s in range(start, end, step):
        # if cp == 1:
        #     p = max(s - k, 0)
        # else:
        #     p = max(k - s, 0)
        p = max(cp * (s - k), 0) * direction
        
        x.append(s)
        y.append(p)
        
    plt.plot(x, y)

plot_option_price(3500, 3000, 5000, 10, 1, -1)
```

```python
def plot_option_price(k=3000, start=2500, end=4000, step=100, cp=1, direction=1, premium=40):
    x = []
    y = []
    
    for s in range(start, end, step):
        # if cp == 1:
        #     p = max(s - k, 0)
        # else:
        #     p = max(k - s, 0)
        p = (max(cp * (s - k), 0) - premium) * direction
        
        x.append(s)
        y.append(p)
        
    plt.plot(x, y)

plot_option_price(3500, 3000, 5000, 10, 1, -1, 200)
```
